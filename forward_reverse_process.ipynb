```python
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll
```

```python
def plot_data(data):
    plt.scatter(data[:, 0], data[:, 1], alpha=0.1)
    plt.xlim([-2,2])
    plt.ylim([-2,2])
```

```python
n_samples = 10_000
data, _ = make_swiss_roll(n_samples, random_state=42)
# operations to make it equal to the paper
data = data[:, [2,0]] / 10
# flip the image vertically by inverting the rows in reverse order
data = data*np.array([1, -1])
plot_data(data)
```

```python
np.random.seed(42)
def forward_process(data, T, betas):
    for t in range(T):
        beta_t = betas[t]
        mu = data * np.sqrt(1-beta_t)
        std = np.sqrt(beta_t)
        data = mu + np.random.randn(data.shape[0], data.shape[1]) * std # data ~ N(mu, std)
    return data
```

```python
T=40
betas = np.linspace(1e-4, 1e-2, T)
xT = forward_process(data, T,betas)
```

```python
# this is not a standard gaussian
xT.mean(0), xT.std(0)
```

```python
plot_data(xT)
```

### After 40 steps this not a well-behaved standard gaussian. We need better parameters to make it standard

```python
def forward_process(data, T, betas):
    for t in range(T):
        beta_t = betas[t]
        mu = data * torch.sqrt(1-beta_t)
        std = torch.sqrt(beta_t)
        # we sample from q(x_t | x_{t-1})
        data = mu + torch.randn_like(data) * std # data ~ N(mu, std)
    return data
```

```python
# let's use some known data. We fix beta from the beginning but it can be learned via gradient descent to make the final
# image follows the standard normal distribution
T=40
betas = torch.sigmoid(torch.linspace(-18, 10, T)) * (3e-1 - 1e-5) + 1e-5
print(data[:2])
xT = forward_process(torch.from_numpy(data), T, betas)
```

```python
# this is a standard gaussian
xT.mean(0), xT.std(0)
```

```python
plot_data(xT)
```

```python
class DiffusionModel():
    def __init__(self, T):
        self.betas = torch.sigmoid(torch.linspace(-18, 10, T)) * (3e-1 - 1e-5) + 1e-5
        self.aphas = 1 - self.betas
        self.alphas_bar = torch.cumprod(self.aphas, dim=0)

    def forward_process(self, x0, t):
        """
        param t: number of diffusion steps
        """
        assert t > 0, "should be greater than zero"
        t = t - 1 # index start at zero
        mu = torch.sqrt(self.alphas_bar[t]) * x0
        std = torch.sqrt(1 - self.alphas_bar[t])
        epsilon = torch.randn_like(x0)
        
        return  mu + epsilon * std # data ~ N(mu, std)
```

```python
model = DiffusionModel(40)
x0 = torch.from_numpy(data)
xT = model.forward_process(x0, 40)
```

```python
# this is a standard gaussian
xT.mean(0), xT.std(0)
```

```python
plot_data(xT)
```

```python
def sample_batch(batch_size, device = "cpu"):
    data, _ = make_swiss_roll(batch_size, random_state=42)
    # operations to make it equal to the paper
    data = data[:, [2,0]] / 10
    # flip the image vertically by inverting the rows in reverse order
    data = data*np.array([1, -1])
    return torch.from_numpy(data).to(device)
```

```python
model = DiffusionModel(40)
x0 =sample_batch(5_000)
xT = model.forward_process(x0, 20)
```

```python
# this is a standard gaussian
xT.mean(0), xT.std(0)
```

```python
class DiffusionModel():
    def __init__(self, T, model: nn.Module):
        self.betas = torch.sigmoid(torch.linspace(-18, 10, T)) * (3e-1 - 1e-5) + 1e-5
        self.aphas = 1 - self.betas
        self.alphas_bar = torch.cumprod(self.aphas, dim=0)
        self.T = T
        self.model = model

    def forward_process(self, x0, t):
        """
        param t: number of diffusion steps
        """
        assert t > 0, "should be greater than zero"
        t <= self.T, f"t should be lower or equal than {self.T}" 
        
        t = t - 1 # index start at zero
        mu = torch.sqrt(self.alphas_bar[t]) * x0
        std = torch.sqrt(1 - self.alphas_bar[t])
        epsilon = torch.randn_like(x0)
        
        return  mu + epsilon * std # data ~ N(mu, std)

    def reverse_process(self,xt, t):
        assert t > 0, "should be greater than zero"
        t <= self.T, f"t should be lower or equal than {self.T}"

        t = t - 1 # index start at zero
        mu, std = self.model(xt, t)
        epsilon = torch.randn_like(x0)
        return  mu + epsilon * std # data ~ N(mu, std)  
```
